based on [yao8839836/text_gcn](<https://github.com/yao8839836/text_gcn>). We use the same `build_graph.py` and `remove_words`
codes from the implementation but create the neural network by pytorch instead of TensorFlow.

## Running training and evaluation

Run the following steps on each of the datasets `20ng`, `R8`, `R52`, `ohsumed` or `mr`.
1. `cd preprocess`
2. `python remove_words.py <dataset>`
3. `python build_graph.py <dataset>`
4. `cd ..`
5. `python fit_model.py <dataset>`


The preprocessing modules are same as [yao8839836/text_gcn](<https://github.com/yao8839836/text_gcn>) and only the model has been rewritten using `pytorch`. The preprocessing time of the datasets are given in the following order with $t_{dataset}$ meaning the time it takes for preprocessing.

$$t_{mr} \le t_{R8} \le t_{ohsumed} \le t_{R52} \le t_{20ng}$$
